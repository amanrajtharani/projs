```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
salary_data = pd.read_csv('ds_salaries.csv')
```

```python
salary_data.info()
```

```python
salary_data.isnull().sum()
```

```python
def categorize_salary(salary):
    if salary > 100000:
        return "High income"
    elif salary == 100000:
        return "Medium income"
    else:
        return "Low income"

salary_data["sal>100000"] = salary_data["salary"].map(categorize_salary)


salary_data
```

```python
plt.figure(figsize=(20,10))
graph = sns.countplot(x="job_title",data=salary_data,hue="sal>100000")
plt.title("Salaries by Job Title and Salary Greater than 100000")
plt.show()
```
